#Machine Learing Assignment

1) What is a parameter?

-> A parameter is a value inside a model that is learned from the data during training (e.g., weights in linear regression).

2) What is correlation?

-> Correlation measures the strength and direction of the relationship between two variables.

3) What does negative correlation mean?

-> Negative correlation means that when one variable increases, the other decreases.

4) Define Machine Learning. What are the main components in Machine Learning?

-> Machine Learning is a field where computers learn patterns from data to make predictions.
Main components: Dataset, Features, Model, Training, Evaluation.

5) How does loss value help in determining whether the model is good or not?

-> Lower loss value means the model’s predictions are closer to the actual values, indicating better performance.

6) What are continuous and categorical variables?

-> Continuous variables: Numeric values with infinite possibilities (e.g., height).
-> Categorical variables: Non-numeric labels or categories (e.g., gender).

7) How do we handle categorical variables in Machine Learning? What are the common techniques?

-> By converting categories into numeric form.
Common techniques: Label Encoding, One-Hot Encoding, Target Encoding.

8) What do you mean by training and testing a dataset?

-> Training: Used to teach the model.
-> Testing: Used to evaluate the model’s performance on unseen data.

9) What is sklearn.preprocessing?

-> A Scikit-learn module that provides tools for scaling, encoding, and transforming data.

10) What is a Test set?

-> A separate portion of the dataset used only for evaluating the trained model.

11) How do we split data for model fitting (training and testing) in Python?

-> Using train_test_split() from sklearn.model_selection.

12) How do you approach a Machine Learning problem?

-> Steps: Understand problem → Collect data → Clean data → EDA → Feature engineering → Model training → Evaluation → Deployment.

13) Why do we have to perform EDA before fitting a model to the data?

-> EDA helps understand data patterns, detect missing values, outliers, and relationships necessary for better model building.

14) What is correlation?

-> Correlation measures the relationship and dependency between two variables.

15) What does negative correlation mean?

-> It means when one variable goes up, the other goes down.

16) How can you find correlation between variables in Python?

-> Using df.corr(), seaborn heatmap, or numpy.corrcoef().

17) What is causation? Explain difference between correlation and causation with an example.

-> Causation means one variable directly affects another.
Difference:
Correlation ≠ Causation.
Example: Ice cream sales and drowning cases are correlated, but ice cream does not cause drowning — hot weather causes both.

18) What is an Optimizer? What are different types of optimizers? Explain each with an example.

-> An optimizer updates model parameters to reduce loss.
Types:

SGD: Updates weights using gradient of each batch.

Adam: Adaptive learning rate + momentum.

RMSprop: Uses moving average of gradients for faster convergence.

19) What is sklearn.linear_model?

-> A module containing linear models like LinearRegression, LogisticRegression, Ridge, Lasso, etc.

-> Trains the model on given data.
Arguments: X (features), y (target).

21) What does model.predict() do? What arguments must be given?

-> Predicts output using the trained model.
Arguments: X_test (input features).

23) What is feature scaling? How does it help in Machine Learning?

-> Scaling transforms features to a similar range.
It improves model performance and training speed, especially for distance-based algorithms.

24) How do we perform scaling in Python?

-> Using StandardScaler, MinMaxScaler from sklearn.preprocessing.

25) Explain data encoding.

-> Data encoding converts categorical values into numeric format so that ML models can understand them (e.g., One-Hot Encoding, Label Encoding).
